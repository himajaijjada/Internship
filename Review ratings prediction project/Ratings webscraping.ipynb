# Collecting reviews and ratings from amazon for washing machine

```python
#Importing libraries reqiured
import selenium
import pandas as pd
import time
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
import requests
import re
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException, ElementClickInterceptedException
from selenium.webdriver.support.ui import WebDriverWait
```

```python
#Connect to web driver
driver = webdriver.Chrome("chromedriver.exe")

#defining URL
url = 'https://www.amazon.in/'

#Getting the website to driver
driver.get(url)
#When we run this line, automatically the webpage will be opened

time.sleep(2) #adding sleep time of 2 seconds

#Searching for washing machine in the search bar 
#Extracting the elements from the Search bar
search_bar = driver.find_element_by_id('twotabsearchtextbox')
search_bar

#Entering 'washing machine' in the search bar
search_bar.send_keys('washing machine')

time.sleep(2) #adding sleep time of 2 seconds

#Extracting the element of search option
search_btn = driver.find_element_by_id('nav-search-submit-button')
search_btn

#clicking on the search button
search_btn.click()
```

```python
#empty list for stoing all URLs of smartphone products
produrl =[]
```

```python
produrl_tag = driver.find_elements_by_xpath('//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
```

```python
#the url of each smartphone is appended to the the list
for u in produrl_tag:
    try: 
        produrl.append(u.get_attribute("href"))
    except:
        produrl.append('-')
```

```python
# an empty list to store all the urls of the comment pages
rateurl = []
```

```python
"""collects all the urls of 1 star - 5 star comment pages.
This is done to ensure that all comments that are being collected, represent each of the 5 different star ratings equally"""
for i in produrl:
    driver.get(i)
    time.sleep(3)
    url_tag5 = driver.find_element_by_xpath('//a[@class="a-link-normal 5star"]')
    url_tag4 = driver.find_element_by_xpath('//a[@class="a-link-normal 4star"]')
    url_tag3 = driver.find_element_by_xpath('//a[@class="a-link-normal 3star"]')
    url_tag2 = driver.find_element_by_xpath('//a[@class="a-link-normal 2star"]')
    url_tag1 = driver.find_element_by_xpath('//a[@class="a-link-normal 1star"]')
    rateurl.append(url_tag5.get_attribute("href"))
    rateurl.append(url_tag4.get_attribute("href"))
    rateurl.append(url_tag3.get_attribute("href"))
    rateurl.append(url_tag2.get_attribute("href"))
    rateurl.append(url_tag1.get_attribute("href"))
```

```python
cmt = [] #empty list to store all comments
rate = [] #empty list to store all the ratings
```

```python
#all the comments and ratings on 5 pages are collected
for i in rateurl:
    driver.get(i)
    for x in range(0,5):
        time.sleep(3)
        comment_tag = driver.find_elements_by_xpath('//span[@class="a-size-base review-text review-text-content"]')
        rate_tag = driver.find_elements_by_xpath('//a[contains(@href,"/gp/customer-reviews/")]/i/span[1]')
        for k in comment_tag:
            try:
                cmt.append(k.text.replace("\n"," "))
            except:
                cmt.append("-")
        for r in rate_tag:
            try:
                rate.append(r.get_attribute("innerHTML"))
            except:
                rate.append("-")
        try:
            nxtbtn = driver.find_element_by_xpath('//*[@id="cm_cr-pagination_bar"]/ul/li[2]')
            nxtbtn.click()
        except NoSuchElementException:
            pass
        except ElementClickInterceptedException:
            pass
        except StaleElementReferenceException:
            pass
```

```python
#removes the "out of 5 stars" text and stores only the numbers in the Rate list
Rate=[]
for r in rate:
    Rate.append(r.replace(' out of 5 stars',''))
```

```python
#creating a dataframe to store the comments and ratings
ratingsdf = {'Comment':cmt,'Rating':Rate}

Ratedf1  = pd.DataFrame(ratingsdf)
```

```python
Ratedf1
```

```python
#Closing the driver
driver.close()
```

# Collecting reviews and ratings from flipkart for 'laptops', 'Phones','smart watches','Monitors'

```python
# Connecting to the webdriver
driver = webdriver.Chrome("chromedriver.exe")
```

```python
url = "https://www.flipkart.com/"
driver.get(url)
time.sleep(2)
#clicking on cancel buttopn 
driver.find_element_by_xpath('//div[@class="_2QfC02"]/button').click()
```

```python
srch_items = ['laptop', 'smart phone','smart watch','Monitor']
```

```python
review_text = []
ratings = []
```

```python
def scrap():    
        for i in driver.find_elements_by_xpath('//div[@class="t-ZTKy"]'):
            review_text.append(i.text)
            
        for i in driver.find_elements_by_xpath('//p[@class="_2-N8zT"]/../div'):
            ratings.append(i.text)
        return
```

```python
urls=[]
```

```python
for i in srch_items:
    #Find search bar
    srchBar = driver.find_element_by_xpath('//div[@class="_3OO5Xc"]/input')
    srchBar.clear()
    srchBar.send_keys(i)
    #clicking on search button
    driver.find_element_by_xpath('//button[@class="L0Z3Pu"]').click()
    time.sleep(3)
    page = []
    for i in driver.find_elements_by_xpath('//nav[@class="yFHi8N"]/a'):
        page.append(i.get_attribute('href'))
    for i in page[0:4]:
        driver.get(i)
        time.sleep(3)
        items = driver.find_elements_by_xpath('//a[@class="_1fQZEK"]')
        for i in items:
            urls.append(i.get_attribute('href'))
```

```python
len(urls)
```

```python
for i in urls:
    driver.get(i)
    time.sleep(2)
    for _ in range(2):
        driver.execute_script("window.scrollBy(0,6000)")
        time.sleep(1)
        #clicking on all reviews
    try:
        btn=driver.find_element_by_xpath('//div[@class="_2c2kV-"]/following::a')
        lnk = btn.get_attribute('href')
        driver.get(lnk)
        time.sleep(1)
    except NoSuchElementException:
        pass

    scrap()        
    try:
        n_page=driver.find_elements_by_xpath('//nav[@class="yFHi8N"]/a')
        np=[]
        for i in n_page:
            np.append(i.get_attribute('href'))
        for i in np[0:18]:
            driver.get(i)
            time.sleep(1)
            scrap()
    except: continue
```

```python
#creating a dataframe
data = list(zip(review_text,ratings))
df2 = pd.DataFrame(data, columns = ["Comment","Rating"])
df2
```
