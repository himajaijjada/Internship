1) Write a python program to display all the header tags from wikipedia.org.

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page

page.content

soup=BeautifulSoup(page.content)
soup

header_tags=soup.find_all('span',class_="mw-headline")
header_tags

Header=[]
for i in header_tags:
    Header.append(i.text.replace('\xa0...',''))
    
Header

import pandas as pd
tags=pd.DataFrame(Header)
tags
```

2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
page=requests.get("https://www.imdb.com/list/ls091520106/")
page

page.content

soup=BeautifulSoup(page.content)
soup

movie_name=soup.find_all('h3',class_="lister-item-header")

Name=[]
for i in movie_name:
    Name.append(i.text.replace('\n',''))

year_of_release=soup.find_all('span',class_="lister-item-year text-muted unbold")
    
Year=[]
for i in year_of_release:
    Year.append(i.text)
    
movie_rating=soup.find_all('div',class_="ipl-rating-star small")

Rating=[]
for i in movie_rating:
    Rating.append(i.text.replace('\n',''))

import pandas as pd
imdb_top100_df = pd.DataFrame({'Movie':Name,'Year_of_release':Year,'Rating':Rating})  # --> creating a dataframe

imdb_top100_df
```

```python
page=requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
page

page.content

soup=BeautifulSoup(page.content,'html.parser')
soup

movie_name=soup.find_all('td',class_="titleColumn")

Name=[]
for i in movie_name:
    Name.append(i.text.replace('\n','').split()[1])

year_of_release=soup.find_all('span',class_="secondaryInfo")
    
Year=[]
for i in year_of_release:
    Year.append(i.text)

movie_rating=soup.find_all('td',class_="ratingColumn imdbRating")

Rating=[]
for i in movie_rating:
    Rating.append(i.text.replace('\n',''))

import pandas as pd
imdb_top100_df = pd.DataFrame({'Movie':Name,'Year_of_release':Year,'Rating':Rating})  # --> creating a dataframe

imdb_top100_df.head(100)
```

7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content
and the code for the video from the link for the youtube video from the post.

```python
page=requests.get("https://coreyms.com/")
page

soup=BeautifulSoup(page.content)
soup

heading=soup.find_all('h2',class_="entry-title")
Heading=[]
for i in heading:
    Heading.append(i.text)
    
Heading

date=soup.find_all('time',class_="entry-time")
Date=[]
for i in date:
    Date.append(i.text)
    
Date

content=soup.find_all('div',class_="entry-content")
Content=[]
for i in content:
    Content.append(i.text)
    
Content

code=soup.find_all('iframe',class_="youtube-player")
Code=[]
for i in code:
    Code.append(i['src'])

Code
Code.insert(4,"No youtube link")
Code

import pandas as pd
post=pd.DataFrame({'Heading':Heading,'Date':Date,'Content':Content,'Code':Code})
post
```
