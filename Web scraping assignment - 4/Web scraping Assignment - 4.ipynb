You need to find following details:

A) Rank

B) Name

C) Artist

D) Upload date

E) Views


```python
#Connect to web driver
driver = webdriver.Chrome("chromedriver.exe")

#defining URL
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'

#Getting the website to driver
driver.get(url)
#When we run this line, automatically the webpage will be opened

#Adding sleeptime
time.sleep(2)


#Connect to web driver
driver = webdriver.Chrome("chromedriver.exe")

#defining URL
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'

#Getting the website to driver
driver.get(url)
#When we run this line, automatically the webpage will be opened

#Adding sleeptime
time.sleep(2) 

#Creating the empty lists to store the scraped data
Rank=[]
Name=[]
Artist=[]
Upload_Date=[]
Views=[]


#As we need only the first 30 details, we will iterate only for first 30 data
#Scrapping the details of the Rank of the video
rank=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[1]")
for i in rank[:30]:
    Rank.append(i.text)
    
#Scrapping the details of the video name
video=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[2]")
for i in video[:30]:
    Name.append(i.text)
    
#Scrapping the details of the Artist name
artist=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[3]")
for i in artist[:30]:
    Artist.append(i.text)

#Scrapping the details of the upload date
date=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[5]")
for i in date[:30]:
    Upload_Date.append(i.text)    
    
#Scrapping the details of the views information
views=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[4]")
for i in views[:30]:
    Views.append(i.text)
    
df=pd.DataFrame({"Rank":Rank,
                 "Name":Name,
                "Uploader/artist":Artist,
                "Date of upload":Upload_Date,
                "Views(in billion)":Views})
df
```

```python
#Removing the stray numbers from videoname
New_name =df["Name"].str.split("[", n = 1, expand = True) 
New_name
```

```python
#Dropping the column with stray numbers
df.drop(columns=['Name'],axis=1,inplace=True)

#Inserting the name column
df.insert(1,"Name",New_name[0])

# To set index as the "Rank" column
df.set_index('Rank',inplace=True)

#Checking the data after removing the stray numbers ans setting the Rank column as index
df  
```

```python
#Closing the driver
driver.close()
```

#### 2. Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.

You need to find following details:

A) Match title (I.e. 1st ODI)

B) Series

C) Place

D) Date

E) Time

Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 


time.sleep(2)
# Opening the bcci.tv
url = "https://www.bcci.tv/"
driver.get(url)

#Adding sleep time
time.sleep(2)

# Accepting cookies to close popup
#Extracting the elements of acceptcookies button
popup = driver.find_element_by_xpath('/html/body/div[15]/button')
#Clicking on accept cookies
popup.click()

#Adding sleep time
time.sleep(2)


#Scrolling to the bottom of the page 
driver.execute_script("window.scrollBy(0,5000)")

#Extracting the elements for the international option
International = driver.find_element_by_xpath('/html/body/div[14]/div/div[1]/div[2]/ul[1]/li[2]/a')

#Clicking on the International option
International.click()


#Adding sleep time
time.sleep(5)

# make empty lists
match=[] 
series=[]
place=[]
date=[]
Time=[]

#scarping match title
                    
match_title = driver.find_elements_by_xpath('//span[@class="matchOrderText ng-binding ng-scope"]')

for i in match_title:
    try :
        match.append(i.text.replace('-',''))
    except NoSuchElementException:#handling no such element exception
        match.append('No details available')

#Adding sleep time
time.sleep(2)

#scarping series title    
series_title = driver.find_elements_by_xpath('//div[@class="fixture-card-mid d-flex align-items-center justify-content-between"]')

for i in series_title:
    try:
        series.append(i.text.replace('\n',''))
    except NoSuchElementException:#handling no such element exception
        series.append("No details Available")

#Adding sleep time
time.sleep(2)

#scarping match venue    
place_title = driver.find_elements_by_xpath('//div[@class="fix-place ng-binding ng-scope"]')
    
for i in place_title:
    try:    
        place.append(i.text.split()[3:])
    except NoSuchElementException:#handling no such element exception
        place.append("No details Available")
        

#Adding sleep time        
time.sleep(2)

#scarping date of match 
date_title = driver.find_elements_by_xpath('//div[@class="match-card-left match-schedule"]')

for i in date_title:
    try:
        date.append(i.text)
    except NoSuchElementException:#handling no such element exception
        date.append("No details Available")      

#Adding sleep time
time.sleep(2)


#scarping time of the match    
time_title = driver.find_elements_by_xpath('//h5[@class="text-right ng-binding"]')
for i in time_title:
    try:    
        Time.append(i.text)
    except NoSuchElementException:#handling no such element exception
        Time.append("No details Available")    

#Adding sleep time
time.sleep(2)


#preparing the dataframe
Cricket_matches=pd.DataFrame({"Match":match,
                "Series":series,
                "Place":place,
                "Date":date,
                "Time":Time})

Cricket_matches
```

```python
#Closing the driver
driver.close()
```

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(2)


# Opening the github
url = "https://github.com/"
driver.get(url)

#Adding sleeptime
time.sleep(3)


# clicking on option button
explore=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details')
explore.click()

#Adding sleeptime
time.sleep(2)

#finding trending option
explore=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a')
try:
    explore.click()
except:
    driver.get(explore.get_attribute('href'))

#Adding sleeptime    
time.sleep(2)

# making empty lists    
repo_urls=[]
Contributors_count=[]
Language_used=[]
repo_title=[]
repo_desc=[]

#scraping repositories urls
repos=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for i in repos:
    repo_urls.append(i.get_attribute('href'))

#Adding sleeptime    
time.sleep(2) 

#scraping repositories title
repo_title=[]
try:
    repos=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']")
    for i in repos:
        repo_title.append(i.text)
except NoSuchElementException:#handling no such element exception
    repo_title.append('No details available')

#Adding sleeptime    
time.sleep(3)
    

for i in repo_urls:
    driver.get(i)
    l=[]
    time.sleep(3)
    #scraping repositories discription
    try:
        repo = driver.find_element_by_xpath('/html/body/div[4]/div/main/div[2]/div/div/div[3]/div[2]/div/div[1]/div/p')
        repo_desc.append(repo.text)
    except NoSuchElementException:
        repo_desc.append('-')
    
    #scraping contributors count
    try:
        count=driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        Contributors_count.append(count.text)
    except NoSuchElementException:#handling no such element exception
        Contributors_count.append('No details available')
    
    #Adding sleeptime
    time.sleep(2)
    
     #scraping languages used   
    languages=driver.find_elements_by_xpath("//li[@class='d-inline']//a//span[1]")
    if languages:    
        for i in languages:
            l.append(i.text)
    else:
        l.append('No languages used')
    Language_used.append(l)   
    
#Adding sleeptime
time.sleep(2)   

#creating dataframe
df=pd.DataFrame({'Title':repo_title,
                'Description':repo_desc,
                'Contributors_count':Contributors_count,
                'Language_used':Language_used})
df
```

#### 7. Scrape the details of Data science recruiters from naukri.com.

B) Designation

C) Company

D) Skills they hire for

E) Location

Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and click on search. All this should be done through code


```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 

#Adding sleeptime
time.sleep(2)

# Opening the naukri.com
url = "https://www.naukri.com/hr-recruiters-consultants"
driver.get(url)

#Adding sleeptime
time.sleep(2)

#accesing search bar
search=driver.find_element_by_xpath("//input[@class='sugInp']")
search.send_keys("Data Science")

#Adding sleeptime
time.sleep(1)

#clicking on search button
button=driver.find_element_by_xpath("//button[@id='qsbFormBtn']")
button.click()

#Adding sleeptime
time.sleep(1)

#scrapping names
name=[]
try:
    names=driver.find_elements_by_xpath("//div[@class='recInfo']/div/p/a[1]")
    for i in names:
        name.append(i.text)
except NoSuchElementException:           #handling no such element exception
    name.append('No details available')

#Adding sleeptime    
time.sleep(2)


#scrapping designation
designation=[]
try:
    designation_list=driver.find_elements_by_xpath("//div[@class='recInfo']/div/p/span[1]")
    for i in designation_list:
        designation.append(i.text)
except NoSuchElementException:        #handling no such element exception
    designation.append('No details available')

#Adding sleeptime    
time.sleep(2)

#scrapping company names
company=[]
try:
    companies=driver.find_elements_by_xpath("//div[@class='recInfo']/div/p/a[2]")
    for i in companies:
        company.append(i.text)
except NoSuchElementException:#handling no such element exception
    company.append('No details available')

#Adding sleeptime    
time.sleep(2)

#scrapping location names
location=[]
soup=BeautifulSoup(driver.page_source,'html.parser')
l=soup.find_all('div',attrs={'vcard'})
for i in l:
    try:
        location.append(i.find('small').text)
    except AttributeError :
        location.append("No details available")

#Adding sleeptime        
time.sleep(2)

#skills they hire for
skill=[]
try:
    skills=driver.find_elements_by_xpath("//div[@class='recInfo']/div[2]")
    for i in skills:
        skill.append(i.text)
except NoSuchElementException:        #handling no such element exception
    skill.append('No details available')

#Adding sleeptime    
time.sleep(2)    
        
#creating dataframe
df=pd.DataFrame({'Name':name,
                'Designation':designation,
                'Company':company,
                "Location":location,
                'Hire_for_Skillset':skill})
df
```
